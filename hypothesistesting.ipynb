One-Sample t-Test:

```python
from scipy.stats import ttest_1samp
import numpy as np
ages = np.array([45, 89, 23, 46, 12, 69, 45, 24, 34, 67])
pop_mean = 50
t_stat, p_value = ttest_1samp(ages, pop_mean)
print("t-statistic:" , t_stat)
print("p-value:" , p_value)
```

Two-Sample t-Test

Compares the means of two independent samples.

```python
from scipy.stats import ttest_ind
sample1 = np.array([29, 30, 33, 41, 38, 36, 35, 31, 29, 30])
sample2 = np.array([31, 32, 33, 39, 30, 33, 30, 28, 29, 31])
t_stat, p_value = ttest_ind(sample1, sample2)
print("t-statistic:" , t_stat)
print("p-value:" , p_value)
```

Paired t-Test

Compares the means of two paired samples(repeated measures)

```python
from scipy.stats import ttest_rel
before = np.array([5, 8, 12, 15, 18])
after = np.array([7, 10, 13, 16, 19])
t_stat, p_value = ttest_rel(before, after)
print("t-statistic:" , t_stat)
print("p-value:" , p_value)
```

ANOVA (F-Test):

Compares the means of three or more groups.

```python
import statsmodels.formula.api as smf # Import the correct module
import pandas as pd
import statsmodels.api as sm # Import statsmodels.api for anova_lm
data = pd.DataFrame({'group' : ['A', 'A', 'B', 'B', 'C', 'C'], 'value': [10, 15,20,25,30,35]})
model = smf.ols('value ~ C(group)' , data=data).fit() #Use smf.ols
aov_table = sm.stats.anova_lm(model, typ=2) #Use sm.stats.anova_lm
print(aov_table)
```

Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
y = np.array([5, 20, 14, 32, 22, 38])
model = LinearRegression()
model.fit(x,y)
print("Coefficients:" , model.coef_, model.intercept_)
y_pred = model.predict(x)
plt.scatter(x, y, color= 'blue' , label= 'Actual data')
plt.plot(x, y_pred, color='red' , label='Regression line')
plt.xlabel('Independent variable (x)')
plt.ylabel('Dependent variaBLE (y)')
plt.title('Linear Regression')
plt.legend()
plt.show()
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y, y_pred)
r2= r2_score(y, y_pred)

print("Mean Squared Error:", mse)
print("R- squared:", r2)
```
